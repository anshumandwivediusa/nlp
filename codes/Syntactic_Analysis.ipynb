# Syntactic Analysis in NLP

## Overview
Syntactic Analysis, also known as Parsing, is the process of analyzing the grammatical structure of a sentence. It determines how words are organized and how they relate to one another according to the rules of a language.

Unlike Lexical Analysis, which focuses on individual words, syntactic analysis examines the relationships between words to understand the sentence's structure.

This notebook demonstrates syntactic analysis techniques using Python libraries.

## Input Example
We will use the following sentence:

> John submitted the project report to Microsoft yesterday.

````python
text = "John submitted the project report to Microsoft yesterday."
````

## 1. Part-of-Speech (POS) Tagging

POS tagging assigns grammatical categories (noun, verb, adjective, etc.) to each word.

````python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp(text)
[(token.text, token.pos_) for token in doc]
````

### Example Output
| Word      | POS Tag     |
|-----------|-------------|
| John      | PROPN       |
| submitted | VERB        |
| the       | DET         |
| project   | NOUN        |
| report    | NOUN        |
| to        | ADP         |
| Microsoft | PROPN       |
| yesterday | ADV         |

## 2. Dependency Parsing

Dependency parsing identifies grammatical relationships between words.

````python
for token in doc:
    print(token.text, "-->", token.dep_, "-->", token.head.text)
````

### Example Output
```
John --> nsubj --> submitted
submitted --> ROOT --> submitted
report --> dobj --> submitted
project --> compound --> report
to --> prep --> submitted
Microsoft --> pobj --> to
yesterday --> npadvmod --> submitted

```

## Explanation of Each Relation

- **John → nsubj → submitted**  
  - *John* is the **nominal subject (nsubj)** of the verb *submitted*.  
  - It tells us *John* is the one performing the action.

- **submitted → ROOT → submitted**  
  - *submitted* is the **root** of the sentence.  
  - Every dependency tree has a root — usually the main verb.

- **report → dobj → submitted**  
  - *report* is the **direct object (dobj)** of *submitted*.  
  - It’s what was submitted.

- **project → compound → report**  
  - *project* is a **compound modifier** of *report*.  
  - Together they form the noun phrase *project report*.

- **to → prep → submitted**  
  - *to* is a **prepositional modifier (prep)** attached to *submitted*.  
  - It introduces the prepositional phrase *to Microsoft*.

- **Microsoft → pobj → to**  
  - *Microsoft* is the **object of the preposition (pobj)** *to*.  
  - It completes the phrase *to Microsoft*.

- **yesterday → npadvmod → submitted**  
  - *yesterday* is a **noun phrase adverbial modifier (npadvmod)** of *submitted*.  
  - It specifies when the action happened.


## 🌟 Putting It Together
The dependency parse tells us:

- **Subject**: John  
- **Verb (Root)**: submitted  
- **Object**: project report  
- **Prepositional Phrase**: to Microsoft  
- **Adverbial Modifier (Time)**: yesterday  

So the structure is:  
**John (subject) → submitted (verb) → project report (object) → to Microsoft (prepositional phrase) → yesterday (time modifier).**

## 3. Constituency Parsing (NLTK)

Constituency parsing divides a sentence into hierarchical phrases.

````python
import nltk
from nltk import CFG

nltk.download('punkt')

# Simple grammar for demonstration
grammar = CFG.fromstring("""
S -> NP VP
NP -> 'John' | 'the' N
N -> 'project' | 'report'
VP -> V NP PP | V NP
V -> 'submitted'
PP -> P NP
P -> 'to'
NP -> 'Microsoft' | 'yesterday'
""")

parser = nltk.ChartParser(grammar)
sentence = ['John', 'submitted', 'the', 'project', 'report', 'to', 'Microsoft']
for tree in parser.parse(sentence):
    tree.pretty_print()
    tree.draw()
````

## 4. Chunking (Shallow Parsing)

Chunking groups nearby words into meaningful phrases.

````python
import nltk

# Ensure resources are available
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')

text = "John submitted the project report to Microsoft yesterday."

# Tokenize and POS tag
tokens = nltk.word_tokenize(text)
pos_tags = nltk.pos_tag(tokens)

# Define a chunk grammar (not CFG!)
chunk_grammar = r"""
  NP: {<DT>?<JJ>*<NN>}   # Noun Phrase
  VP: {<VB.*><NP|PP>*}   # Verb Phrase
  PP: {<IN><NP>}         # Prepositional Phrase
"""

# Apply chunk parser
cp = nltk.RegexpParser(chunk_grammar)
result = cp.parse(pos_tags)

# Visualize
# Print the chunk tree in text format
print(result)

# Pretty-print the tree structure
result.pretty_print()

````

## Summary

- **POS Tagging** assigns grammatical categories.
- **Dependency Parsing** identifies word relationships.
- **Constituency Parsing** builds hierarchical phrase structures.
- **Chunking** detects major phrase boundaries.

Together, these techniques provide structural understanding for semantic analysis and downstream NLP tasks.
