# Lexical Analysis in NLP

Lexical meaning is the meaning of an individual word as defined in a dictionary (lexicon).

**Lexical Analysis** is the process of analyzing raw text at the **word (lexical) level**. It converts unstructured text into meaningful lexical units (tokens) while preparing the text for syntactic and semantic analysis.

In a typical NLP pipeline, lexical analysis comes **after basic text cleaning** and **before syntactic analysis**.

## Techniques in Lexical Analysis

| Technique | Purpose |
|-----------|---------|
| Tokenization | Split text into tokens (words, subwords, punctuation). |
| Text Normalization | Standardize text (lowercase, expand contractions, etc.). |
| Stop Word Removal | Remove common words that add little meaning. |
| Stemming | Reduce words to their root form using rules. |
| Lemmatization | Convert words to their dictionary form. |
| Lexicon Lookup | Match words against a dictionary or vocabulary. |
| Handling OOV | Manage words not found in the vocabulary. |

```python
# Install required libraries (uncomment if running locally)
# !pip install nltk spacy
# !python -m spacy download en_core_web_sm8

import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
import spacy

nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
text = "hello world"
```

## Example Input
We will use the sentence:

> "Microsoft announced the launch of Copilot for Microsoft 365 in Seattle on 15 March 2025."

```python
text = "Microsoft announced the launch of Copilot for Microsoft 365 in Seattle on 15 March 2025."
```

### Tokenization

```python
tokens = word_tokenize(text)
tokens
```

### Normalization (lowercasing)

```python
normalized = [t.lower() for t in tokens]
normalized
```

### Stopword Removal

```python
stop_words = set(stopwords.words('english'))
filtered = [t for t in normalized if t not in stop_words]
filtered
```

### Stemming

```python
stemmer = PorterStemmer()
stemmed = [stemmer.stem(t) for t in filtered]
stemmed
```

### Lemmatization

```python
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(t) for t in filtered]
lemmatized
```

### Lexicon Lookup & OOV Handling (using spaCy)

```python
doc = nlp(text)
[(token.text, token.is_oov) for token in doc]
```

## Difference Between Text Preprocessing and Lexical Analysis

| Text Preprocessing | Lexical Analysis |
|--------------------|------------------|
| Cleans and standardizes raw text. | Analyzes text at the word/token level. |
| Removes noise (HTML, URLs, emojis, punctuation). | Produces lexical units (tokens) for further analysis. |
| Often includes cleaning and normalization. | Focuses on vocabulary, token formation, and word normalization. |
| Goal: Prepare text for NLP. | Goal: Convert text into meaningful lexical units. |
