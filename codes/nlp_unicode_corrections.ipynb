```python
import unicodedata
```

```python
examples = [
    ("New Line", "\n", "Hello\nWorld"),
    ("Carriage Return", "\r", "Hello\rWorld"),
    ("Tab", "\t", "Hello\tWorld"),
    ("Form Feed", "\f", "Hello\fWorld"),
    ("Vertical Tab", "\v", "Hello\vWorld"),
    ("Null Character", "\0", "Hello\0World"),
    ("Zero Width Space", "\u200B", "Hello\u200BWorld"),
    ("Zero Width Non-Joiner", "\u200C", "Hello\u200CWorld"),
    ("Zero Width Joiner", "\u200D", "Hello\u200DWorld"),
    ("Byte Order Mark (BOM)", "\uFEFF", "\uFEFFHello World"),
    ("Left-to-Right Mark", "\u200E", "Hello\u200EWorld"),
    ("Right-to-Left Mark", "\u200F", "Hello\u200FWorld"),
]

for name, char, text in examples:
    print("=" * 60)
    print(f"{name}")
    print(f"Unicode : U+{ord(char):04X}")
    print(f"repr()  : {repr(text)}")
    print(f"Display : {text}")
```
