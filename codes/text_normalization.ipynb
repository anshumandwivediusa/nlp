# Text Normalization - Full Working Examples

```
import re, unicodedata
text="I'm HAPPY!!! Visit https://Example.com on 15th March, 2025. Café Cafe\u0301 Sooooo coool!!!"
print(text)
```

```
# Case Folding
print(text.lower())
```

```
contractions={"i'm":"i am","don't":"do not","can't":"cannot","it's":"it is"}
t=text.lower()
for k,v in contractions.items(): t=t.replace(k,v)
print(t)
```

```
s='Hello   \t\n World'
print(repr(re.sub(r'\\s+',' ',s).strip()))
```

```
s='Hello!!! Wait..... Really????'
s=re.sub(r'!{2,}','!',s);s=re.sub(r'\\?{2,}','?',s);s=re.sub(r'\\.{2,}','.',s);print(s)
```

```
s='Revenue 1,000,000 on 15th March.'
s=re.sub(r'(?<=\\d),(?=\\d)','',s);s=re.sub(r'(\\d+)(st|nd|rd|th)',r'\\1',s);print(s)
```

```
a='Café';b='Cafe\u0301';print(a==b);print(unicodedata.normalize('NFC',a)==unicodedata.normalize('NFC',b))
```

```
print(re.sub(r'(.)\\1{2,}',r'\\1\\1','Sooooooo coooool!!!!'))
```

```
t=unicodedata.normalize('NFC',text).lower()
for k,v in contractions.items(): t=t.replace(k,v)
t=re.sub(r'!{2,}','!',t)
t=re.sub(r'(?<=\\d),(?=\\d)','',t)
t=re.sub(r'(\\d+)(st|nd|rd|th)',r'\\1',t)
t=re.sub(r'(.)\\1{2,}',r'\\1\\1',t)
t=re.sub(r'\\s+',' ',t).strip()
print(t)
```
